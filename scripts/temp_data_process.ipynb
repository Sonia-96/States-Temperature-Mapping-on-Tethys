```python
import pandas as pd
import json
```

```python
data_folder = '../tethysapp/state_temp_map/workspaces/app_workspace/states_temp'
```

```python

file_path = f'{data_folder}/input/temperature.json'
with open(file_path, 'r') as file:
    json_data = json.load(file)
```

```python
data_dict = {}
for entry in json_data:
    state_name = entry['state']
    temp_data  = entry['data']
    state_temp_data = {data['year']: data['temperature'] for data in temp_data}
    data_dict[state_name] = state_temp_data
df = pd.DataFrame.from_dict(data_dict)
df.index.name='Year'
df
```

```python
output_path = f"{data_folder}/output/"
```

```python
for col in df.columns:
    df_state = df[col].dropna()
    df_state.to_csv(f"{output_path}/{col}.csv",)
```
